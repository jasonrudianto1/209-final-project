```python
def export_team(team_id):
    print("exporting team " + str(team_id))
    df = teamyearbyyearstats.TeamYearByYearStats(team_id=team_id).get_data_frames()[0]
    df.to_csv(working_dir + str(team_id) + '.csv', index=False)
    seen.add(team_id)
```

```python
for team in all_team_ids:
    if team in seen:
        continue
    export_team(team)
print("finished")
```

```python
def read_df(team_id):
    path = working_dir + str(team_id) + '.csv'
    return pd.read_csv(path)
```
